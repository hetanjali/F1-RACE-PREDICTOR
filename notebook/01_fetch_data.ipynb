# 01 - Data Fetching
Fetch 2023 F1 race results using FastF1 and save to the `/data` folder.

```python
import fastf1
print("fastf1 ok")
import pandas as pd
print("pandas ok")
import os
print("os ok")
```

```python
# Create cache folder if it doesn't exist
os.makedirs('../data/cache', exist_ok=True)

# Tell FastF1 to use it
fastf1.Cache.enable_cache('../data/cache')
```

```python
session = fastf1.get_session(2023,'bahrain','R')
session.load()
```

```python
results = session.results
print(results.shape)
results.head(20)
```

```python
columns_to_keep = [
    'DriverNumber', 'BroadcastName', 'FullName',
    'TeamName', 'Position', 'Points', 'Status',
    'GridPosition', 'Time'
]
df = results[columns_to_keep].copy()
df.head(20)
```

```python
df['year'] = 2023
df['RaceName']= 'Bahrain'
df['Round']= 1

df.head(20)
```

```python
df['Top3'] = (df['Position'] <= 3).astype(int)

print(df[['FullName', 'Position', 'Top3']])
```

```python
os.makedirs('../data', exist_ok=True)

df.to_csv('../data/bahrain_2023.csv', index=False)

print("Saved! ✅")
```

 # Fetch multiple races in a loop

```python
races = [
    (1, 'Bahrain'), (2, 'Saudi Arabia'), (3, 'Australia'),
    (4, 'Azerbaijan'), (5, 'Miami'), (6, 'Monaco'),
    (7, 'Spain'), (8, 'Canada'), (9, 'Austria'),
    (10, 'Britain'), (11, 'Hungary'), (12, 'Belgium'),
    (13, 'Netherlands'), (14, 'Italy'), (15, 'Singapore'),
    (16, 'Japan'), (17, 'Qatar'), (18, 'United States'),
    (19, 'Mexico'), (20, 'Brazil'), (21, 'Las Vegas'),
    (22, 'Abu Dhabi')
]

all_dfs = []

for round_num, race_name in races:
    print(f"Fetching {race_name}...")
    try:
        s = fastf1.get_session(2023, race_name, 'R')
        s.load(telemetry=False, weather=False, messages=False)
        r = s.results[columns_to_keep].copy()
        r['Year'] = 2023
        r['RaceName'] = race_name
        r['Round'] = round_num
        r['Top3'] = (r['Position'] <= 3).astype(int)
        all_dfs.append(r)
    except Exception as e:
        print(f"  ⚠️ Skipped {race_name}: {e}")

full_df = pd.concat(all_dfs, ignore_index=True)
print(f"\nTotal rows: {full_df.shape[0]}")
full_df.head()
```

```python
full_df.to_csv('../data/f1_2023_all_races.csv', index=False)
print(f"Full dataset saved! Shape: {full_df.shape} ✅")
```
